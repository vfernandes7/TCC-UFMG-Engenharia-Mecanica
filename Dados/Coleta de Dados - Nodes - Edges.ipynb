```python
for index_areas_analisadas in range(len(areas_analisadas)):
    
    limite_superior = areas_analisadas[index_areas_analisadas][0]
    limite_inferior = areas_analisadas[index_areas_analisadas][1]
    limite_esquerdo = areas_analisadas[index_areas_analisadas][2]
    limite_direito  = areas_analisadas[index_areas_analisadas][3]
    
    G = ox.graph_from_bbox(limite_superior,limite_inferior,limite_esquerdo,limite_direito,network_type='walk')

    fig, ax = ox.plot_graph(G)
    
    nodes, edges = ox.graph_to_gdfs(G, nodes=True, edges=True)
    
    print('nodes',len(nodes))
    print('edges',len(edges))
    
    nodes.to_excel('nodes_loop_'+str(index_areas_analisadas)+'.xlsx')
    edges.to_excel('edges_loop_'+str(index_areas_analisadas)+'.xlsx')
```
